```python
import pandas as pd
import numpy as np

np.random.seed(42)

n_users = 5000

data = pd.DataFrame({
    "user_id": range(1, n_users + 1),
    "installed": 1,
    "tutorial_complete": np.random.binomial(1, 0.85, n_users),
})

data["level1_complete"] = np.where(
    data["tutorial_complete"] == 1,
    np.random.binomial(1, 0.70, n_users),
    0
)

data["first_purchase"] = np.where(
    data["level1_complete"] == 1,
    np.random.binomial(1, 0.25, n_users),
    0
)

data["revenue"] = data["first_purchase"] * np.random.gamma(2, 10, n_users)

data.head()
```

```python
total_users = len(data)

tutorial_rate = data["tutorial_complete"].mean()
level1_rate = data["level1_complete"].mean()
purchase_rate = data["first_purchase"].mean()

print("Tutorial Conversion:", round(tutorial_rate, 2))
print("Level 1 Conversion:", round(level1_rate, 2))
print("Purchase Conversion:", round(purchase_rate, 2))
```

```python
drop_tutorial = 1 - tutorial_rate
drop_level1 = tutorial_rate - level1_rate
drop_purchase = level1_rate - purchase_rate

print("Drop after Install:", round(drop_tutorial, 2))
print("Drop after Tutorial:", round(drop_level1, 2))
print("Drop after Level1:", round(drop_purchase, 2))
```

```python
total_revenue = data["revenue"].sum()
arpu = total_revenue / total_users

print("Total Revenue:", round(total_revenue, 2))
print("ARPU:", round(arpu, 2))
```

```python
import matplotlib.pyplot as plt

stages = [
    total_users,
    data["tutorial_complete"].sum(),
    data["level1_complete"].sum(),
    data["first_purchase"].sum()
]

labels = ["Install", "Tutorial", "Level 1", "Purchase"]

plt.figure()
plt.plot(labels, stages)
plt.title("Funnel Analysis")
plt.xlabel("Stage")
plt.ylabel("Number of Users")
plt.show()
```

```python
high_spenders = data[data["revenue"] > data["revenue"].mean()]

print("High spender ratio:", round(len(high_spenders)/len(data),2))
```

```python
plt.figure()
plt.hist(data["revenue"], bins=30)
plt.title("Revenue Distribution")
plt.xlabel("Revenue")
plt.ylabel("User Count")
plt.show()
```

```python
paying_users = data[data["first_purchase"] == 1]

print("Average revenue per paying user:", round(paying_users["revenue"].mean(),2))
```

```python
data["segment"] = np.where(
    data["revenue"] > data["revenue"].median(),
    "High Value",
    "Low Value"
)

data.head()
```

```python
import matplotlib.pyplot as plt

segments = data["segment"].unique()

for segment_name in segments:
    segment_data = data[data["segment"] == segment_name]

    stages_segment = [
        len(segment_data),
        segment_data["tutorial_complete"].sum(),
        segment_data["level1_complete"].sum(),
        segment_data["first_purchase"].sum()
    ]

    labels = ["Install", "Tutorial", "Level 1", "Purchase"]

    plt.figure()
    plt.plot(labels, stages_segment)
    plt.title(f"Funnel Analysis for {segment_name} Users")
    plt.xlabel("Stage")
    plt.ylabel("Number of Users")
    plt.show()
```

```python
data["segment"] = np.where(
    data["first_purchase"] == 1,
    "Paying Users",
    "Non-Paying Users"
)
```

```python
segment_funnel = data.groupby("segment").agg(
    users=("user_id", "count"),
    tutorial_rate=("tutorial_complete", "mean"),
    level1_rate=("level1_complete", "mean"),
    purchase_rate=("first_purchase", "mean")
)

segment_funnel
```
